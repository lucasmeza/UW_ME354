### Calculate the mean, standard deviation and standard error of a dataset

```python
def mean(X):
    return sum(X)/len(X)

def stdDev(X):
    return (sum([(x-mean(X))**2 for x in X])/(len(X)))**0.5

def stdErr(X):
    return stdDev(X)/len(X)**0.5

# Example
A = [1,2,3,4,5,6]
print(mean(A))
print(stdDev(A))
print(stdErr(A))
```

### Calculate the uncertainty of an equation using the max/min bound

```python
def funct(x,y):
    return x*y

def bound(f,x,y,xmax,ymax,xmin,ymin):
    return [f(x,y), (f(xmax,ymax)-f(xmin,ymin))/2]

# Example
a = 5; sa = 0.2
b = 10; sb = 0.3
print(bound(funct, a, b, a+sa, b+sb, a-sa, b-sb))
```

### Use Python functions to do linear regressions

```python
from scipy.stats import linregress #this is the inbuilt linear regression function in scipy

#Example
from random import uniform
X = [1,2,3,4,5,6]
Y = [x+2.2*uniform(-1,1) for x in X]
XY = linregress(X,Y)
print(XY)
```

```python
#Plot the results
import matplotlib.pyplot as plt

#Linear fit
f = [XY.slope*x + XY.intercept for x in X]

fig,ax = plt.subplots(1)
ax.plot(X,Y,'.',label='Data')
ax.plot(X,f,'--',label='Fit ='+str(round(XY.slope,2)))
ax.legend()
```
