# 3-point vs 4-point Bending Moment and Deflection 

## Define some variables

```python
L = 1 #beam length
a = L/4 #4-point bend position
n = 100 #number of points for plotting
X = [i*L/n for i in range(n)] #bar positions
M3 = []; M4 = []; w3 = []; w4 = [] #empty variables
```

## Calculate all the moments and deflections

```python
for x in X:
    #3-point Bending
    if x<L/2:
        M3 += [x/2]
        w3 += [3*L**2*x-4*x**3]
    else:
        M3 += [(L-x)/2]
        w3 += [4*x**3-12*L*x**2+9*x*L**2-L**3]
    
    #4-point Bending
    if x<a:
        M4 += [x/2]
        w4 += [x*(12*a*L-4*x**2-12*a**2)]
    elif x>=a and x<(L-a):
        M4 += [a/2]
        w4 += [a*(12*L*x-4*a**2-12*x**2)]
    else:
        M4 += [(L-x)/2]
        w4 += [4*(L-x)*(3*a*L-(L-x)**2-3*a**2)]
```

## Plot the results

### Plot bending moment

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams.update({'font.size': 20})

fig = plt.figure(figsize=(10,5))
ax = fig.gca()
ax.plot(X,M3,label='$M_3$')
ax.plot(X,M4,label='$M_4$')
ax.set_xlabel('x position')
ax.set_ylabel('Moment (M/P)')
ax.legend();
```

### Plot deflection

```python
fig = plt.figure(figsize=(10,5))
ax = fig.gca()
ax.plot(X,w3,label='$w_3$')
ax.plot(X,w4,label='$w_4$')
ax.set_xlabel('x position')
ax.set_ylabel('Deflection (w/(P/48EI)')
ax.legend();
```
